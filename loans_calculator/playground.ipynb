```python
def calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity):
    """
    Calculates the loan payment for a given periodicity.

    Parameters:
    - interest_rate (float): Annual interest rate as a percentage.
    - num_payments (int): Total number of payments.
    - original_balance (float): The loan amount.
    - periodicity (int): Number of payments per year (e.g., 12 for monthly, 6 for bimonthly, 4 for quarterly, 1 for yearly).

    Returns:
    - float: The calculated loan payment.
    """
    annual_interest_rate = interest_rate / 100
    periodic_interest_rate = annual_interest_rate / periodicity
    discount_factor = ((1 + periodic_interest_rate) ** num_payments - 1) / (periodic_interest_rate * (1 + periodic_interest_rate) ** num_payments)

    payment = original_balance / discount_factor
    return payment

# Example usage:
interest_rate = 5.0  # 5% annual interest rate
num_payments = 12  # 12 payments per year for monthly
original_balance = 10000.0  # loan amount

monthly_payment = calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity=12)
bimonthly_payment = calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity=6)
quarterly_payment = calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity=4)
yearly_payment = calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity=1)

print(f"Monthly Payment: {monthly_payment}")
print(f"Bimonthly Payment: {bimonthly_payment}")
print(f"Quarterly Payment: {quarterly_payment}")
print(f"Yearly Payment: {yearly_payment}")
```

```python
import pandas as pd

# Loan details
interest_rate = 9.0  # 9% annual interest rate
num_payments = 5 * 12  # 5 years with semi-annual payments
original_balance = 10_000_000.0  # 10 million USD
periodicity = 12  # Semi-annual payments

# Calculate the semi-annual payment
semi_annual_payment = calculate_custom_period_payment(interest_rate, num_payments, original_balance, periodicity)

# Initialize lists to store cash flow components
outstanding_balance = [original_balance]
interest_paid = []
principal_paid = []

# Calculate cash flow for each period
for period in range(1, num_payments + 1):
    interest = outstanding_balance[-1] * (interest_rate / 100 / periodicity)
    principal = semi_annual_payment - interest
    outstanding_balance.append(outstanding_balance[-1] - principal)
    interest_paid.append(interest)
    principal_paid.append(principal)

# Create a DataFrame
cash_flow_df = pd.DataFrame({
    'Period': list(range(1, num_payments + 1)),  # Adjusted the range to match the number of periods
    'Outstanding Balance': outstanding_balance[:-1],  # Exclude the last value added for the next period
    'Interest Paid': interest_paid,
    'Principal Paid': principal_paid
})

# Display the DataFrame
cash_flow_df.to_clipboard()

```
